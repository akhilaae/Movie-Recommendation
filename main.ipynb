```python
import pandas as pd
```

```python
data=pd.read_csv("dataset.csv")
```

```python
data
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data=data[["id","title","genre","overview"]]
```

```python
data
```

```python
data["tags"]=data["genre"]+data["overview"]
```

```python
data
```

```python
data=data[["id","title","tags"]]
```

```python
data
```

```python
data.isnull().sum()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000,stop_words="english")
vector=cv.fit_transform(data["tags"].values.astype("U")).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
distance = sorted(list(enumerate(similarity[2])),reverse=True,key=lambda x:x[1])
for i in distance[1:6]:
        print(data.iloc[i[0]].title)
```

```python
def recommend(movie):
    index=data[data["title"]==movie].index[0]
    distance = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distance[1:6]:
        print(data.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
